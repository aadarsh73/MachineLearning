```python
import numpy as np
```

```python
def gradient_descent(x,y):
  w_curr = b_curr = 0
  iterations = 10000
  m = len(x)
  rate = 0.1
  for i in range(iterations):
    y_predicted = w_curr * x + b_curr
    cost = (1/2*m)*sum([val**2 for val in (y_predicted-y)])
    wd = (1/m)*sum(x*(y_predicted-y))
    bd = (1/m)*sum(y_predicted-y)
    w_curr = w_curr - rate*wd
    b_curr = b_curr - rate*bd
    print( "w {}, b {}, cost {}, iteration {}".format(w_curr,b_curr,cost,i))
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)
```
